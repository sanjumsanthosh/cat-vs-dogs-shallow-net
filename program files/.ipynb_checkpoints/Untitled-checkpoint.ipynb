```python
import numpy as np
import cv2
import os
import glob
import pandas as pd
from sklearn.preprocessing import LabelBinarizer
```

```python
csv_location = '/Users/sanjaymsanthosh/Desktop/projects/project 2019/cat vs dog classifier/dataset/train_values.csv'
train_dataset_location = '/Users/sanjaymsanthosh/Desktop/projects/project 2019/cat vs dog classifier/dataset/train'
```

```python
for(_,_,filename) in os.walk(train_dataset_location):break
filename = np.array(filename)
```

```python
def get_data_at_loction(location):
    sample_image_loc = os.path.sep.join([train_dataset_location,filename[location]])
    return(cv2.imread(sample_image_loc))
```

```python
def show_image_at_location(location):
    sample_image = cv2.imshow(filename[location],get_data_at_loction(location))
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

```python
def preprocess(source,dim=(32,32)):
    source = cv2.cvtColor(source,cv2.COLOR_BGR2GRAY)
    return cv2.resize(source,dim,cv2.INTER_AREA)
```

```python
def preprocess_to_hex():
    out = []
    for images in range(filename.shape[0]/10):
        img = get_data_at_loction(images)
        img_formated = preprocess(img)
        reshaped = np.reshape(img_formated,[1,1024])
        rows = np.append(reshaped,filename[images].split('.')[0])
        out.append(rows)
        if(images%500)==1:print(images/25000)
    return out
```
